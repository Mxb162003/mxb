```r

library(ggplot2)


set.seed(42)
x <- rnorm(1000)
y <- rnorm(1000)
data <- data.frame(x, y)


p <- ggplot(data, aes(x, y)) +
     stat_density2d(aes(fill = after_stat(level)), geom = "polygon") +
     scale_fill_viridis_c() +
     theme_minimal() +
     labs(title = "Heatmap de densitÃ© de points", x = "X", y = "Y")


ggsave("heatmap.png", plot = p)


print(p)

```
